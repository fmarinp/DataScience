```python
df['Gender'] = df['Sex'].map( {'female': 0, 'male': 1} ).astype(int)
```

```python
df.Embarked[pd.isnull(df.Embarked)==True] = 'N'
```

```python
df['Embarked_index'] = df['Embarked'].map( {'S': 0, 'C': 1,'Q': 2,'N':3} ).astype(int)
```

```python
median_ages = np.zeros((2,3))
median_ages
```

```python
for i in range(0, 2):
    for j in range(0, 3):
        median_ages[i,j] = df[(df['Gender'] == i) & \
                              (df['Pclass'] == j+1)]['Age'].dropna().median()
 
median_ages
```

```python
df['AgeFill'] = df['Age']

df[ df['Age'].isnull() ][['Gender','Pclass','Age','AgeFill']].head(10)
```

```python
for i in range(0, 2):
    for j in range(0, 3):
        df.loc[ (df.Age.isnull()) & (df.Gender == i) & (df.Pclass == j+1),\
                'AgeFill'] = median_ages[i,j]
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch']
df['Age*Class'] = df.AgeFill * df.Pclass
```

```python
df = df.drop(['Name', 'Sex', 'Ticket', 'Cabin', 'Embarked','Embarked_index'df = df.drop(['Age'], axis=1)
 
```

```python
#df = df.drop(['Age'], axis=1)
df = df.dropna()
df = df.drop(['PassengerId','SibSp','Parch','AgeFill','FamilySize','Age*Class'],axis=1)
train_data = df.values
train_data
list(df)
```

```python
# Import the random forest package
from sklearn.ensemble import RandomForestClassifier 

# Create the random forest object which will include all the parameters
# for the fit
forest = RandomForestClassifier(n_estimators = 100)
```

```python
# Fit the training data to the Survived labels and create the decision trees
forest = forest.fit(train_data[0::,1::],train_data[0::,0])
```

```python
# All the ages with no data -> make the median of all Ages
median_age = test_df['Age'].dropna().median()
if len(test_df.Age[ test_df.Age.isnull() ]) > 0:
    test_df.loc[ (test_df.Age.isnull()), 'Age'] = median_age

# All the missing Fares -> assume median of their respective class
if len(test_df.Fare[ test_df.Fare.isnull() ]) > 0:
    median_fare = np.zeros(3)
    for f in range(0,3):                                              # loop 0 to 2
        median_fare[f] = test_df[ test_df.Pclass == f+1 ]['Fare'].dropna().median()
    for f in range(0,3):                                              # loop 0 to 2
        test_df.loc[ (test_df.Fare.isnull()) & (test_df.Pclass == f+1 ), 'Fare'] = median_fare[f]

# Collect the test data's PassengerIds before dropping it
ids = test_df['PassengerId'].values
```
